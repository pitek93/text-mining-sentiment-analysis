```python
from wordcloud import WordCloud
```

```python
text = open("train/pos/0_9.txt", encoding='utf8').read()
```

```python
import matplotlib.pyplot as plt
```

```python
# wc = WordCloud()
wc = WordCloud(width=1000, height=595, prefer_horizontal=0.0)
wc.generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```
